```python
from seleniumwire import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json

def log_request(request, file):
    if request.response and request.response.status_code < 400:
        log_entry = {
            "url": request.url,
            "method": request.method,
            "status_code": request.response.status_code,
            "request_headers": dict(request.headers),
            "response_headers": dict(request.response.headers),
            "response_body": request.response.body.decode('utf-8', errors='replace')
        }
        json.dump(log_entry, file)
        file.write('\n')  # Add a newline for readability

# Set up the Chrome WebDriver with selenium-wire
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)

# URL of the webpage you want to interact with
# url = 'https://ec.toranoana.jp/tora_r/ec/cot/ranking/daily/all/'  # Uncomment and replace if needed
url = 'https://meeymap.com/'
url = 'https://remaps.vn/maps/my-maps'

try:
    with open('network_log.jsonl', 'w', encoding='utf-8') as log_file:
        driver.get(url)

        # Wait for the page to load
        time.sleep(10)
        del driver.requests

        while True:
            print("Time to scroll")

            # Scroll down
            # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait for any AJAX requests to complete
            time.sleep(10)
            print("Logging requests...")
            
            # Log all successful request URLs, headers, and response contents
            for request in driver.requests:
                log_request(request, log_file)

            del driver.requests

except KeyboardInterrupt:
    print("Stopping traffic capture...")    

finally:
    # Close the browser
    driver.quit()
    print("Browser closed. Logging complete.")
```

```python
import json
from pprint import pprint

def load_first_request(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        first_line = file.readline().strip()
        if first_line:
            return json.loads(first_line)
        else:
            return None

def display_request(request):
    if request is None:
        print("No requests found in the log file.")
        return

    print("URL:", request['url'])
    print("Method:", request['method'])
    print("Status Code:", request['status_code'])
    
    print("\nRequest Headers:")
    pprint(request['request_headers'])
    
    print("\nResponse Headers:")
    pprint(request['response_headers'])
    
    print("\nResponse Body (first 500 characters):")
    print(request['response_body'][:500] + "..." if len(request['response_body']) > 500 else request['response_body'])

# Load and display the first request
file_path = 'network_log.jsonl'

first_request = load_first_request(file_path)
display_request(first_request)
```
