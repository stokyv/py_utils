```python
from shapely.geometry import Polygon, Point
import matplotlib.pyplot as plt

# List of lat/lon coordinates
coordinates = [
    {'lat': 21.02234129, 'lng': 105.83045434},
    {'lat': 21.022338, 'lng': 105.83046377},
    {'lat': 21.02230455, 'lng': 105.83045083},
    {'lat': 21.02231743, 'lng': 105.83041416},
    {'lat': 21.02231854, 'lng': 105.83041108},
    {'lat': 21.022381, 'lng': 105.83043488},
    {'lat': 21.02237098, 'lng': 105.83046477},
    {'lat': 21.02234129, 'lng': 105.83045434}
]

# Extract (lon, lat) pairs from the list of dictionaries
# Note: Shapely uses (x, y) order, which corresponds to (longitude, latitude)
polygon_points = [(coord['lng'], coord['lat']) for coord in coordinates]

# Create a Polygon object
polygon = Polygon(polygon_points)

# Print some information about the polygon
print(f"Polygon is valid: {polygon.is_valid}")
print(f"Polygon area: {polygon.area}")
print(f"Polygon centroid: {polygon.centroid.x}, {polygon.centroid.y}")

# Plotting the polygon
x, y = polygon.exterior.xy
plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.fill(x, y, alpha=0.3)
plt.title("Polygon from Lat/Lon Coordinates")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.axis('equal')
plt.grid(True)

# Add points to the plot
for coord in coordinates:
    plt.plot(coord['lng'], coord['lat'], 'ro')  # 'ro' means red circle marker

plt.tight_layout()
plt.show()

# Example: Check if a point is inside the polygon
test_point = Point(105.83043, 21.02235)
print(f"Is point {test_point.x}, {test_point.y} inside the polygon? {polygon.contains(test_point)}")
```

```python
import matplotlib.pyplot as plt
from shapely.geometry import Polygon

# List of coordinates
coordinates = [
    {'lat': 21.02234129, 'lng': 105.83045434},
    {'lat': 21.022338, 'lng': 105.83046377},
    {'lat': 21.02230455, 'lng': 105.83045083},
    {'lat': 21.02231743, 'lng': 105.83041416},
    {'lat': 21.02231854, 'lng': 105.83041108},
    {'lat': 21.022381, 'lng': 105.83043488},
    {'lat': 21.02237098, 'lng': 105.83046477},
    {'lat': 21.02234129, 'lng': 105.83045434}
]

coordinates = [
    {'lat': 10.78616219341603, 'lng': 106.7001832518976},
    {'lat': 10.786132636483948, 'lng': 106.70015225443636},
    {'lat': 10.78615140998346, 'lng': 106.70013265508283},
    {'lat': 10.78618241727456, 'lng': 106.70016237702286},
    {'lat': 10.78616219341603, 'lng': 106.7001832518976},
]

# Extracting the (lng, lat) tuples
points = [(point['lng'], point['lat']) for point in coordinates]

# Creating the polygon
polygon = Polygon(points)

# Finding the centroid of the polygon
centroid = polygon.centroid
centroid_x, centroid_y = centroid.x, centroid.y

# Plotting the polygon
x, y = polygon.exterior.xy
plt.figure()
plt.plot(x, y)
plt.fill(x, y, alpha=0.5, fc='r', ec='black')

# Plotting the centroid
plt.plot(centroid_x, centroid_y, 'bo')  # 'bo' means blue color and circle marker
plt.text(centroid_x, centroid_y, '  Centroid', horizontalalignment='left', verticalalignment='bottom')

plt.title('Polygon with Centroid')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Printing the centroid coordinates
print(f"Centroid: Latitude, Longitude: {centroid_y}, {centroid_x}")
```
